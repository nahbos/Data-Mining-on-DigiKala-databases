```python
s3 = pd.Series([1, 3, 2, np.nan], name='My numeric values')
```

```python
a = pv.validate_numeric(
     s3,
     nullable=False,
     unique=True,
     integer=True,
     min_value=1,
     max_value=2,
     return_type=None)

print(a)
```
